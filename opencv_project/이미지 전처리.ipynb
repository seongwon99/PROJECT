### 이미지 전처리

```python
dstList1=[]
for name in file_names:
    # 1. grayscale
    imgData=cv2.imread(file_path2+name,cv2.IMREAD_GRAYSCALE)
    
    # 2. resize
    imgData=cv2.resize(imgData,dsize=(100,100))
    
    # 3. remove
    output=remove(imgData)

    # 4. 회전
    for degree in range(0,360,10):
        matrix=cv2.getRotationMatrix2D((50,50),degree,1)
        dst=cv2.warpAffine(output,matrix,(100,100))
        dstList1.append(dst)
```

```python
dstList1255=[i/255 for i in dstList1]
dstList1255=[i.flatten() for i in dstList1255]
```

```python
for i in range(len(dstList1255)):
    dstList1255[i]=np.append(dstList1255[i],1)
```

- 고양이상 남자연예인 전처리 과정

```python
file_path = './zico/'
file_path2 = './baek/'
file_path3 = './kang/'
file_path4 = './xiu/'

file_names = os.listdir(file_path3)
file_names
```

```python
dstList2=[]
for name in file_names:
    # 1. grayscale
    imgData=cv2.imread(file_path3+name,cv2.IMREAD_GRAYSCALE)
    
    # 2. resize
    imgData=cv2.resize(imgData,dsize=(100,100))
    
    # 3. remove
    output=remove(imgData)

    # 4. 회전
    for degree in range(0,360,10):
        matrix=cv2.getRotationMatrix2D((50,50),degree,1)
        dst=cv2.warpAffine(output,matrix,(100,100))
        dstList2.append(dst)
```

```python
dstList2255=[i/255 for i in dstList2]
dstList2255=[i.flatten() for i in dstList2255]
```

```python
for i in range(len(dstList2255)):
    dstList2255[i]=np.append(dstList2255[i],1)
```
