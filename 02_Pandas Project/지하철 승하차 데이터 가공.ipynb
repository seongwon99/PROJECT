#### 3-4) 월별 승하차인원 상위,하위역

```python
# 1,2,12월 승차인원 상5
subwaywinter01_on_t=subwaywinter01.sort_values('승차인원 합계',ascending=False).iloc[:5].reset_index(drop=True)
subwaywinter02_on_t=subwaywinter02.sort_values('승차인원 합계',ascending=False).iloc[:5].reset_index(drop=True)
subwaywinter12_on_t=subwaywinter12.sort_values('승차인원 합계',ascending=False).iloc[:5].reset_index(drop=True)

# 1,2,12월 하차인원 상5
subwaywinter01_off_t=subwaywinter01.sort_values('하차인원 합계',ascending=False).iloc[:5].reset_index(drop=True)
subwaywinter02_off_t=subwaywinter02.sort_values('하차인원 합계',ascending=False).iloc[:5].reset_index(drop=True)
subwaywinter12_off_t=subwaywinter12.sort_values('하차인원 합계',ascending=False).iloc[:5].reset_index(drop=True)


# 1,2,12월 승차인원 하5
subwaywinter01_on_w=subwaywinter01.sort_values('승차인원 합계',ascending=False).iloc[-5:].reset_index(drop=True)
subwaywinter02_on_w=subwaywinter02.sort_values('승차인원 합계',ascending=False).iloc[-5:].reset_index(drop=True)
subwaywinter12_on_w=subwaywinter12.sort_values('승차인원 합계',ascending=False).iloc[-5:].reset_index(drop=True)

# 1,2,12월 하차인원 하5
subwaywinter01_off_w=subwaywinter01.sort_values('하차인원 합계',ascending=False).iloc[-5:].reset_index(drop=True)
subwaywinter02_off_w=subwaywinter02.sort_values('하차인원 합계',ascending=False).iloc[-5:].reset_index(drop=True)
subwaywinter12_off_w=subwaywinter12.sort_values('하차인원 합계',ascending=False).iloc[-5:].reset_index(drop=True)
```

```python
subwaywinter01_off_t
```

```python
# 1,2,12월 승차인원 상5에서 하차인원 합계열 제거
subwaywinter01_on_t.drop('하차인원 합계',axis=1,inplace=True)
subwaywinter02_on_t.drop('하차인원 합계',axis=1,inplace=True)
subwaywinter12_on_t.drop('하차인원 합계',axis=1,inplace=True)

# 1,2,12월 승차인원 하5에서 하차인원 합계열 제거
subwaywinter01_on_w.drop('하차인원 합계',axis=1,inplace=True)
subwaywinter02_on_w.drop('하차인원 합계',axis=1,inplace=True)
subwaywinter12_on_w.drop('하차인원 합계',axis=1,inplace=True)


# 1,2,12월 하차인원 상5에서 승차인원 합계열 제거
subwaywinter01_off_t.drop('승차인원 합계',axis=1,inplace=True)
subwaywinter02_off_t.drop('승차인원 합계',axis=1,inplace=True)
subwaywinter12_off_t.drop('승차인원 합계',axis=1,inplace=True)

# 1,2,12월 하차인원 하5에서 승차인원 합계열 제거
subwaywinter01_off_w.drop('승차인원 합계',axis=1,inplace=True)
subwaywinter02_off_w.drop('승차인원 합계',axis=1,inplace=True)
subwaywinter12_off_w.drop('승차인원 합계',axis=1,inplace=True)
```

```python
# 승차인원 합계 / 하차인원 합계 변경 =====> 승차상위 5개 / 하차상위 5개

subwaywinter01_on_t.rename(columns={'승차인원 합계':'승차상위 5개'},inplace=True)
subwaywinter02_on_t.rename(columns={'승차인원 합계':'승차상위 5개'},inplace=True)
subwaywinter12_on_t.rename(columns={'승차인원 합계':'승차상위 5개'},inplace=True)


subwaywinter01_on_w.rename(columns={'승차인원 합계':'승차하위 5개'},inplace=True)
subwaywinter02_on_w.rename(columns={'승차인원 합계':'승차하위 5개'},inplace=True)
subwaywinter12_on_w.rename(columns={'승차인원 합계':'승차하위 5개'},inplace=True)


subwaywinter01_off_t.rename(columns={'하차인원 합계':'하차상위 5개'},inplace=True)
subwaywinter02_off_t.rename(columns={'하차인원 합계':'하차상위 5개'},inplace=True)
subwaywinter12_off_t.rename(columns={'하차인원 합계':'하차상위 5개'},inplace=True)

subwaywinter01_off_w.rename(columns={'하차인원 합계':'하차하위 5개'},inplace=True)
subwaywinter02_off_w.rename(columns={'하차인원 합계':'하차하위 5개'},inplace=True)
subwaywinter12_off_w.rename(columns={'하차인원 합계':'하차하위 5개'},inplace=True)
```

```python
# 1월 자료 합치기 전에 슬라이싱
subwaywinter01_off_t=subwaywinter01_off_t[subwaywinter01_off_t.columns[:]]
subwaywinter01_on_w=subwaywinter01_on_w[subwaywinter01_on_w.columns[:]]
subwaywinter01_off_w=subwaywinter01_off_w[subwaywinter01_off_w.columns[:]]

# 2월 자료 슬라이싱
subwaywinter02_off_t=subwaywinter02_off_t[subwaywinter02_off_t.columns[:]]
subwaywinter02_on_w=subwaywinter02_on_w[subwaywinter02_on_w.columns[:]]
subwaywinter02_off_w=subwaywinter02_off_w[subwaywinter02_off_w.columns[:]]

# 12월 자료 슬라이싱
subwaywinter12_off_t=subwaywinter12_off_t[subwaywinter02_off_t.columns[:]]
subwaywinter12_on_w=subwaywinter12_on_w[subwaywinter02_on_w.columns[:]]
subwaywinter12_off_w=subwaywinter12_off_w[subwaywinter02_off_w.columns[:]]
```

```python
# 1월 자료
subway01=pd.concat([subwaywinter01_on_t,subwaywinter01_off_t,subwaywinter01_on_w,subwaywinter01_off_w],axis=1)

# 2월 자료
subway02=pd.concat([subwaywinter02_on_t,subwaywinter02_off_t,subwaywinter02_on_w,subwaywinter02_off_w],axis=1)

#12월 자료
subway12=pd.concat([subwaywinter12_on_t,subwaywinter12_off_t,subwaywinter12_on_w,subwaywinter12_off_w],axis=1)
```

```python
subway01.insert(0,'사용월',['202201','202201','202201','202201','202201'])
subway02.insert(0,'사용월',['202202','202202','202202','202202','202202'])
subway12.insert(0,'사용월',['202212','202212','202212','202212','202212'])
```

### 5) 월별 혼잡도

```python
con01=subwaywinter01.copy()
con02=subwaywinter02.copy()
con12=subwaywinter12.copy()
```

```python
con01['승하차인원 총합계']=con01['승차인원 합계']+con01['하차인원 합계']
con02['승하차인원 총합계']=con02['승차인원 합계']+con02['하차인원 합계']
con12['승하차인원 총합계']=con12['승차인원 합계']+con12['하차인원 합계']

# count,bin_dividers=np.histogram(subwaywinter01['승하차인원 총합계'],bins=3)
```

```python
name=['여유','보통','혼잡']

con01['교통 혼잡도']=pd.cut(x=con01['승하차인원 총합계'],bins=3,labels=name,include_lowest=True)
con02['교통 혼잡도']=pd.cut(x=con02['승하차인원 총합계'],bins=3,labels=name,include_lowest=True)
con12['교통 혼잡도']=pd.cut(x=con12['승하차인원 총합계'],bins=3,labels=name,include_lowest=True)
```
