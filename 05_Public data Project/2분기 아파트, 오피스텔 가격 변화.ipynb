```python
DF=DF[['시군구','전용면적(㎡)','계약년월','거래금액(만원)','건축년도']]
```

```python
DF=DF.iloc[6:]
```

```python
DF.columns=['구','전용면적','계약년월','거래금액','건축년도']
```

```python
DF['계약년월'] = pd.to_datetime(DF['계약년월'], format='%Y%m')
DF_month = DF[(DF['계약년월'].dt.month >= 5) & (DF['계약년월'].dt.month <= 7)]
```

```python
DF_month['구']=DF_month['구'].apply(lambda x:x.split(' ')[1])
```

```python
DF_month
```

```python
DF_month['거래금액']=DF_month['거래금액'].apply(lambda x:x.replace(',',''))
```

```python
DF_month
```

```python
DF_month['거래금액']=DF_month['거래금액'].apply(lambda x: int(x))
DF_month['전용면적']=DF_month['전용면적'].apply(lambda x: float(x))

a=DF_month.거래금액/DF_month.전용면적

DF_month['1평당가격(만원/㎡)']=a
```

```python
DF_month
```

```python
##################################################
# 건축년도가 2020 ~ 2023 인것을 순살 아파트라고 칭함
# 건축년도가 ~(2020 ~ 2023) 인것을 비순살 아파트라고 칭함
##################################################
```

```python
# trading_df = trading_df[trading_df['시군구']!='군위군']


def find_area_month_avg(data, loc,time):
    area=data[data['구']==loc]

    if time=='순살':
        area=area[area['건축년도'].isin([2020,2021,2022,2023])]
    elif time!='순살':
        area=area[~area['건축년도'].isin([2020,2021,2022,2023])]

    return area

def quarter_contract_count(area):
    
    month_dict={}

    fst=area[area['계약년월'].dt.month==5]
    month_dict['5월']=sum(fst['1평당가격(만원/㎡)'].values.tolist())/len(fst)

    fst=area[area['계약년월'].dt.month==6]
    month_dict['6월']=sum(fst['1평당가격(만원/㎡)'].values.tolist())/len(fst)

    fst=area[area['계약년월'].dt.month==7]
    month_dict['7월']=sum(fst['1평당가격(만원/㎡)'].values.tolist())/len(fst)
    
    return print(month_dict)



loc_list = ['중구','동구','서구','남구','북구','수성구','달서구','달성군']

for i in loc_list:
    quarter_contract_count(find_area_month_avg(DF_month,i,'비순살'))




```
